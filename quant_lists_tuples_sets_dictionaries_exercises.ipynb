#### üìä **Exercise 1: Portfolio Analysis with Lists**

You have the following portfolio data:
- **Stocks**: `["AAPL", "GOOGL", "MSFT", "TSLA"]`
- **Shares**: `[150, 25, 200, 50]`
- **Prices**: `[175.50, 2850.75, 415.25, 890.30]`

**Calculate the total portfolio value and find the largest position.**

**Mathematical Formulas:**
$$\text{Total Value} = \sum_{i=0}^{n-1} \text{shares}[i] \times \text{prices}[i]$$
$$\text{Largest Position} = \max_{i} (\text{shares}[i] \times \text{prices}[i])$$



````python
stocks = ["AAPL", "GOOGL", "MSFT", "TSLA"]
shares = [150, 25, 200, 50]
prices = [175.50, 2875.75, 415.25, 890.30]
positions = []

# Calculating Positions
for i in range(len(shares)):
    position_size = shares[i] * prices[i]
    positions.append(position_size)

# Calculating Total Value
total_value = sum(positions)

# Calculating Largest Position
largest_position = max(positions)

# Print Output
print(f"Total Value: ${total_value:.2f}")
print(f"Largest Position: ${largest_position:.2f}")
````

#### üîó **Exercise 3: Portfolio Overlap with Sets**

You have two portfolios:
- **Portfolio A**: `{"AAPL", "GOOGL", "MSFT", "TSLA"}`
- **Portfolio B**: `{"MSFT", "TSLA", "NVDA", "META"}`

**Find the overlap and unique positions in each portfolio.**

**Set Operations:**
$$\text{Overlap} = A \cap B$$
$$\text{Unique to A} = A - B$$
$$\text{Unique to B} = B - A$$



````python
portfolio_a = {"AAPL", "GOOGL", "MSFT", "TSLA"}
portfolio_b = {"MSFT", "TSLA", "NVDA", "META"}

# Finding the overlap
overlap = portfolio_a & portfolio_b
print(f"Overlap: {overlap}")

# Unique Positions to A
unique_a = portfolio_a - portfolio_b
print(f"Unique to A: {unique_a}")

# Unique Positions to B
unique_b = portfolio_b - portfolio_a
print(f"Unique to B: {unique_b}")
````

#### üìñ **Exercise 4: Portfolio Management with Dictionaries**

Create a portfolio dictionary with the following data:
- **AAPL**: 150 shares at $175.50
- **GOOGL**: 25 shares at $2850.75
- **MSFT**: 200 shares at $415.25

**Calculate total portfolio value and add a new position for TSLA (50 shares at $890.30).**

**Mathematical Formula:**
$$\text{Total Value} = \sum_{i=1}^{n} \text{shares}_i \times \text{price}_i$$



````python
portfolio_dict = {"AAPL" : (150, 175.50), "GOOGL" : (25, 2850.75), "MSFT" : (200, 415.25)}

# Calculating AAPL Position
aapl_position = portfolio_dict.get("AAPL")[0] * portfolio_dict.get("AAPL")[1]

# Calculating GOOGL Position
googl_position = portfolio_dict.get("GOOGL")[0] * portfolio_dict.get("GOOGL")[1]

# Calculating MSFT Position
msft_position = portfolio_dict.get("MSFT")[0] * portfolio_dict.get("MSFT")[1]

# Calculating total portfolio value
portfolio_value = aapl_position + googl_position + msft_position
print(f"Initial Portfolio Value: ${portfolio_value:.2f}")

# Adding TSLA Position
portfolio_dict.update({"TSLA" : (50, 890.30)})
print(f"Updated Portfolio: {portfolio_dict}")

# Calculating TSLA Value
tsla_position = portfolio_dict.get("TSLA")[0] * portfolio_dict.get("TSLA")[1]

# Calculating New Portfolio Value
new_portfolio_value = portfolio_value + tsla_position
print(f"New Portfolio Value: ${new_portfolio_value:.2f}")

````

#### üìà **Exercise 6: Risk Analysis with Lists**

Calculate risk metrics from daily returns:
- **Returns**: `[0.025, -0.012, 0.008, 0.015, -0.003, 0.022, -0.005]`
- **Find**: Maximum gain, maximum loss, and average return

**Mathematical Formulas:**
$$\text{Max Gain} = \max(r_1, r_2, \ldots, r_n)$$
$$\text{Max Loss} = \min(r_1, r_2, \ldots, r_n)$$
$$\text{Average Return} = \frac{1}{n} \sum_{i=1}^{n} r_i$$


````python
daily_returns = [0.025 , - 0.012, 0.008, 0.015, -0.003, 0.022, - 0.005]

# Calculating Max Gain
max_gain = max(daily_returns)
print(f"Max Gain: {(max_gain * 100):.2f}%")

# Calculating Max Loss
max_loss = min(daily_returns)
print(f"Max Loss: {(max_loss * 100):.2f}%")

# Calculating Average Return
avg_return = sum(daily_returns)/len(daily_returns)
print(f"Average Return: {(avg_return*100):.2f}%")
````

#### üéØ **Exercise 8: Portfolio Rebalancing with Sets**

You have a target portfolio allocation:
- **Target Sectors**: `{"Technology": 40%, "Healthcare": 25%, "Finance": 20%, "Energy": 15%}`

Your current portfolio has these sectors:
- **Current Sectors**: `{"Technology", "Healthcare", "Technology", "Finance", "Technology"}`

**Find which sectors need rebalancing and which are missing.**

**Set Operations:**
$$\text{Missing Sectors} = \text{Target Sectors} - \text{Current Sectors}$$
$$\text{Overweight Sectors} = \text{Current Sectors} - \text{Target Sectors}$$

**Code Structure:**
```python
# Convert to sets for comparison
target_set = set(target_sectors.keys())
current_set = set(current_sectors)
missing = target_set - current_set
overweight = current_set - target_set
```

**üí° Hint:** Use set difference operations to find missing and excess sectors.


````python
# Define target sectors with allocations
target_sectors = {
    "Technology": 0.40,
    "Healthcare": 0.25, 
    "Finance": 0.20,
    "Energy": 0.15
}

# Current portfolio sectors
current_sectors = ["Technology", "Healthcare", "Technology", "Finance", "Technology"]

# Convert to sets for comparison
target_set = set(target_sectors.keys())
current_set = set(current_sectors)

# Find missing and overweight sectors
missing_sectors = target_set - current_set
overweight_sectors = current_set - target_set

print("Missing sectors:", missing_sectors)
print("Overweight sectors:", overweight_sectors)

# Calculate current sector allocation
from collections import Counter
current_allocation = Counter(current_sectors)
print("\nCurrent sector allocation:")
for sector, count in current_allocation.items():
    print(f"{sector}: {count/len(current_sectors)*100:.1f}%")
````

#### üîß **Exercise 9: Data Structure Integration**

Combine all data structures to create a complete trading system:
1. **Lists**: Store daily prices `[175.50, 176.25, 175.75, 177.00, 176.50]`
2. **Tuples**: Create trade records `(date, action, symbol, price, quantity)`
3. **Sets**: Track unique sectors `{"Technology", "Healthcare", "Finance"}`
4. **Dictionaries**: Store portfolio data `{symbol: {shares, price, sector}}`

**Create a complete portfolio tracking system with all data structures.**

**System Architecture:**
```python
# Complete trading system structure
class TradingSystem:
    def __init__(self):
        self.price_history = []           # List for time series
        self.trade_records = []           # List of tuples
        self.sectors = set()              # Set for unique sectors
        self.portfolio = {}               # Dictionary for positions
```

**üí° Hint:** Use all four data structures together for comprehensive portfolio management.


````python
class TradingSystem:
    def __init__(self):
        self.price_history = [] # List of dicts
        self.trade_records = [] # List of tuples
        self.sectors = set() # Set for unique sectors in portfolio
        self.portfolio = {} # Dict: Symbol
    
    # Storing Trade
    def add_trade(self,date, action, symbol, price, quantity, sector):

        # Storing Price
        stock_price = {symbol : price }
        self.price_history.append(stock_price)
    
        #Storing Trade Records
        trade = (date, action, symbol, price, quantity)
        self.trade_records.append(trade)

        #Storing Sectors
        self.sectors.add(sector)

        #storing Portfolio
        self.portfolio[symbol] = {'shares': quantity, 'price': price, 'sector': sector}

# Demo

# Create a system
system = TradingSystem()

# Add some trades
system.add_trade("2025-11-15", "BUY", "AAPL", 175.50, 10, "Technology")
system.add_trade("2025-11-16", "BUY", "JNJ", 160.00, 5, "Healthcare")
system.add_trade("2025-11-17", "BUY", "JPM", 140.00, 8, "Finance")
system.add_trade("2025-11-18", "BUY", "NVDA", 4300, 4, "Technology")


# Print output of all the structures
print(f"Price History: {system.price_history}")
print(f"Trade Records: {system.trade_records}")
print(f"Unique Sectors:{system.sectors}")
print(f"Portfolio: {system.portfolio}")
````

#### üí∞ **Exercise 10: Advanced Portfolio Analysis**

Create a comprehensive portfolio dictionary with nested data:
- **AAPL**: `{"shares": 150, "price": 175.50, "sector": "Technology", "beta": 1.2}`
- **GOOGL**: `{"shares": 25, "price": 2850.75, "sector": "Technology", "beta": 1.1}`
- **MSFT**: `{"shares": 200, "price": 415.25, "sector": "Technology", "beta": 0.9}`

**Calculate total portfolio value, sector allocation, and weighted average beta.**

**Mathematical Formulas:**
$$\text{Total Value} = \sum_{i=1}^{n} \text{shares}_i \times \text{price}_i$$
$$\text{Sector Allocation} = \text{Count by sector}$$
$$\text{Weighted Beta} = \frac{\sum_{i=1}^{n} \beta_i \times \text{position\_value}_i}{\text{total\_value}}$$

**Code Structure:**
```python
# Nested dictionary access
for symbol, data in portfolio.items():
    shares = data["shares"]
    price = data["price"]
    sector = data["sector"]
    beta = data["beta"]
```

**üí° Hint:** Use nested dictionary access and list comprehensions.


````python
portfolio = {"AAPL" : {"shares" : 150, "price" : 175.50, "sector" : "Technology", "beta" : 1.2},
             "GOOGL" : {"shares" : 25, "price" : 2850.75, "sector" : "Technology", "beta" : 1.1},
             "MSFT" : {"shares" : 200, "price" : 415.25, "sector" : "Technology", "beta" : 0.9}
             }

# Calculate Total Value
total_value = sum(data["shares"] * data["price"]for data in portfolio.values())
print(f"Total Portfolio Value: ${total_value:.2f}")

# Determine Sector Allocation
sector_counts = {}
for data in portfolio.values():
    sector = data["sector"]
    sector_counts[sector] = sector_counts.get(sector, 0) + 1

print("Sector Allocation:")

for sector, count in sector_counts.items():
    percentage = (count / len(portfolio)) * 100
    print(f"{sector} : {percentage:.2f}%")

# Calculate Weighted Beta
weighted_beta = sum(
    data["beta"] * data["shares"] * data["price"] / total_value
    for data in portfolio.values()
    ) 
print(f"Weighted Beta: {weighted_beta:.2f}")
````
