#### üìä **Exercise 1: Portfolio Valuation**

You own the following positions in your portfolio:
- **150 shares** of AAPL at **$175.50** per share
- **200 shares** of GOOGL at **$2,850.75** per share  
- **100 shares** of MSFT at **$415.25** per share

**Calculate your total portfolio value.**

```
Total Value = (AAPL_shares √ó AAPL_price) + (GOOGL_shares √ó GOOGL_price) + (MSFT_shares √ó MSFT_price)
```

**üí° Hint:** Use multiplication and addition operations.


````python
Portfolio = {"AAPL" : (150, 175.50), "GOOGL" : (200, 2850.75), "MSFT" : (100, 415.25)}

# Calculate position value per ticker
aapl_position = Portfolio.get("AAPL")[0] * Portfolio.get("AAPL")[1]
googl_position = Portfolio.get("GOOGL")[0] * Portfolio.get("GOOGL")[1]
msft_position = Portfolio.get("MSFT")[0] * Portfolio.get("MSFT")[1]

# Calculate total value
total_value = aapl_position + googl_position + msft_position

# Print Output
print(f"Total Value: {total_value}")

````

#### üìà **Exercise 2: Price Change Analysis**

A stock's price moved from **$125.50** to **$138.75**.

**Calculate the absolute price change and determine if this represents a gain or loss.**

```
Price Change = New_Price - Old_Price
Absolute Change = |Price Change|
```

**üí° Hint:** Use subtraction and `abs()` function.


````python
new_price = 125.50
old_price = 138.75

# Calculate change in price and the absolute price change
price_change = (new_price - old_price)
abs_change = abs(price_change)

# Determine if this represents a gain or loss
if price_change < 0:
    print("This represents a Loss")
elif price_change > 0:
    print("This represents a gain")
else:
    print("Price is unchanged")

# Print Output
print(f"Absolute Price Change: {abs_change}")
````

#### üí∞ **Exercise 3: Compound Interest**

Calculate the future value of **$10,000** invested for **5 years** at an annual interest rate of **7.5%** using compound interest.

**Compound Interest Formula:**
```
FV = PV √ó (1 + r)^n
```

Where:
- **FV** = Future Value
- **PV** = Present Value ($10,000)
- **r** = Annual interest rate (0.075)
- **n** = Number of years (5)

**üí° Hint:** Use the `pow()` function for exponentiation.


````python
PV = 10000
r = 0.075
n = 5

# Calculate Future Value
FV = PV * pow((1+r),n)

# Print Output
print(f"Future Value: ${round(FV,2)}")
````

You receive market data as strings:

#### ‚ö†Ô∏è **Exercise 5: Risk Analysis**

Calculate the maximum and minimum returns from these daily returns:
**2.5%, -1.8%, 0.5%, -3.2%, 1.1%**

**Also calculate the average return.**

```
Average Return = (Return1 + Return2 + Return3 + Return4 + Return5) √∑ 5
```

**üí° Hint:** Use `max()`, `min()`, and basic arithmetic operations.


````python
daily_returns = [0.025, -0.018, 0.005, -0.032, 0.011]

# Calculate Max and Min Returns
max_return = max(daily_returns)
min_return = min(daily_returns)

# Calculate Average Return
avg_return = sum(daily_returns)/len(daily_returns)

# Print Output
print(f"Max Return: {(max_return*100):.1f}%")
print(f"Min Return: {min_return*100:.1f}%")
print(f"Average Return: {avg_return*100:.1f}%")
````

#### üßÆ **Exercise 7: Floor Division for Position Sizing**

You have **$10,000** to invest and want to buy shares of a stock priced at **$247.83** per share.

**How many complete shares can you afford? How much cash will remain?**

```
Shares = Cash √∑ Price (floor division)
Remaining Cash = Cash % Price (modulo)
```

**üí° Hint:** Use floor division (`//`) and modulo (`%`) operations.


````python
initial_cash = 10000
stock_price = 247.83

# Calculate quantity of shares purchased
shares = initial_cash // stock_price

# Calculate remaining cash after purchase
remaining_cash = initial_cash % stock_price

# Print Output
print(f"Total Shares Purchased: {int(shares)}")
print(f"Remaining Cash After Purchase: ${remaining_cash:.2f}")
````
